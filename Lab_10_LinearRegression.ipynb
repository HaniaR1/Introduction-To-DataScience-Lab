# Simple Linear Regression
-  Simple linear regression aims to find a linear relationship to describe the correlation between an independent and possibly dependent variable. The regression line can be used to predict or estimate missing values, this is known as interpolation.

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv("BTC-USD.csv")
dataset.dropna(inplace=True)
```

Two variables y and yo are:
- y for the Closing Price
- yo for the Opening Price

```python
X = dataset[['High']]
y = dataset[['Close']]
yo = dataset[['Open']]
```

## Splitting the dataset into the Training set and Test set

The train_test_split function is a powerful tool in Scikit-learn's arsenal, primarily used to divide datasets into training and testing subsets. This function is part of the sklearn. model_selection module, which contains utilities for splitting data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_train, X_test, yo_train, yo_test = train_test_split(X, yo, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

sklearn.linear_model class allows us to fit a linear model to a dataset, predict new values, and evaluate the model's performance. To use the LinearRegression class, we first need to import it from sklearn. linear_model module. We can then create an instance of the class and call its fit method to train the model on a dataset.

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
regressor.fit(X_train, yo_train)
```

## Predicting the Test set results
- predict() : given a trained model, predict the label of a new set of data. This method accepts one argument, the new data X_new (e.g. model. predict(X_new) ), and returns the learned label for each object in the array.

```python
y_pred = regressor.predict(X_test)
y_pred
```

## Visualising the Training set results 

We visualise training set results between High vs Close and High vs Open, using a scatter plot and a line plot. 
- Scatter plot represents the training data 
- Line plot has regressor predict(X train) on the y axis


```python
plt.scatter(X_train['High'], y_train, color = 'black')
plt.plot(X_train['High'], regressor.predict(X_train), color = 'pink')
plt.title('High vs Close (Training set)')
plt.xlabel('High Price')
plt.ylabel('Close Price')
plt.show()
```

```python
plt.scatter(X_train['High'], yo_train, color = 'black')
plt.plot(X_train['High'], regressor.predict(X_train), color = 'cyan')
plt.title('High vs Open (Training set)')
plt.xlabel('High price')
plt.ylabel('Open price')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test['High'], y_test, color = 'black')
plt.plot(X_train['High'], regressor.predict(X_train), color = 'pink')
plt.title('High vs Close (Test set)')
plt.xlabel('High')
plt.ylabel('Close')
plt.show()
```

We see that the plots are close to the straight linear regression line that must mean that our model is working efficiently


```python
plt.scatter(X_test['High'], yo_test, color = 'black')
plt.plot(X_train['High'], regressor.predict(X_train), color = 'pink')
plt.title('High vs Open (Test set)')
plt.xlabel('High')
plt.ylabel('Open')
plt.show()
```
