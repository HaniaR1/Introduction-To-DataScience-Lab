# Kernel Density Estimation (KDE)

- Kernel Density Estimation (KDE) is a technique used for estimating the probability density function (PDF) of a continuous random variable.

### Data Smoothing: 
- KDE smooths out the distribution of data points to estimate the probability density function.

### Kernel Function: 
- It employs a kernel function, typically a probability density function, which acts as a weighting function. Commonly used kernel functions include Gaussian (normal), Epanechnikov, triangular, and rectangular kernels.

### Combination: 
- The individual contributions from all data points are combined to obtain the overall density estimate.

### Bandwidth: 
- KDE also involves selecting a bandwidth parameter, which controls the smoothness of the resulting density estimate. A smaller bandwidth leads to a more detailed (but potentially noisy) estimate, while a larger bandwidth results in a smoother estimate.

```python
# Load Titanic dataset
KDE_data = pd.read_csv("titanic.csv")
KDE_data
```

```python
KDE_data.dropna() #drop all the nullvalues
```

```python
#convert the entire column to a numeric value
KDE_data["boat"] = pd.to_numeric(KDE_data['boat'],  errors='coerce')
KDE_data1 = KDE_data[['boat']]
KDE_data1.dropna()
```

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot KDE plots with different kernels and bandwidths
sns.kdeplot(data = KDE_data1, bw_method=0, kernel='gau', color='blue', fill=True, ax=axes[0, 0])
axes[0, 0].set_title('Gaussian kernel, bandwidth = 0')

sns.kdeplot(data = KDE_data1, bw_method=3, kernel='gau', fill=True, ax=axes[0, 1], color='blue', alpha = 0.4)
axes[0, 1].set_title('Gaussian kernel, bandwidth = 3')

sns.kdeplot(data = KDE_data1, bw_method=6, kernel='gau', color='green', fill=True, ax=axes[1, 0], alpha = 0.7)
axes[1, 0].set_title('Gaussian kernel, bandwidth = 6')

sns.kdeplot(data = KDE_data1, bw_method=10, kernel='gau', color='purple', fill=True, ax=axes[1, 1], alpha = 1)
axes[1, 1].set_title('Gaussian kernel, bandwidth = 10')

# Adjust layout
plt.tight_layout()
plt.show()
```

### 3.	Plot different types of Probability distribution with random values

# Normal Distribution
**A normal distribution, also known as a Gaussian distribution, is a probability distribution that is symmetrical around the mean, with data near the mean occurring more frequently than data further away.**

- loc=0 specifies that the mean of the normal distribution is set to 0. This means that the center of the bell-shaped curve will be at 0.

- scale=1 indicates that the standard deviation of the normal distribution is set to 1. The standard deviation controls the spread or width of the bell curve: larger values of standard deviation result in wider curves, while smaller values result in narrower curves.

- size=1000 specifies that we want to generate a sample of 1000 data points from the normal distribution.

```python
normal_data = np.random.normal(loc=0, scale=1, size=1000)
# Plotting the distributions
plt.figure(figsize=(12, 8))

# Normal Distribution
plt.subplot(2, 2, 1)
sns.histplot(normal_data, kde=True)
plt.title('Normal Distribution')
plt.show()
```

# Uniform Distribution
**Uniform distribution refers to a type of probability distribution in which all outcomes are equally likely.**

- low=0 specifies the lower bound of the uniform distribution. In this case, it's set to 0, meaning that the distribution starts from 0.

- high=1 specifies the upper bound of the uniform distribution. In this case, it's set to 1, meaning that the distribution ends at 1.

```python
uniform_data = np.random.uniform(low=0, high=1, size=1000)

# Uniform Distribution
plt.subplot(2, 2, 2)
sns.histplot(uniform_data, kde=True)
plt.title('Uniform Distribution')
plt.show()
```

# Exponential Distribution
**A continuous probability distribution that describes the time between events in a Poisson point process. In a Poisson point process, events occur continuously and independently at a constant average rate.**

- scale=1 specifies the scale parameter of the exponential distribution. This parameter represents the inverse of the rate parameter (lambda) and determines the mean or average time between events. In this case, it's set to 1, indicating that the average time between events is 1 unit of time.

```python
exponential_data = np.random.exponential(scale=1, size=1000)

# Exponential Distribution
plt.subplot(2, 2, 3)
sns.histplot(exponential_data, kde=True)
plt.title('Exponential Distribution')
plt.show()
```

# Log-normal Distribution

**Is a continuous probability distribution of a random variable where the logarithm is normally distributed.
This distribution is commonly used when dealing with variables that are naturally positive and skewed to the right, such as financial assets or income.**

- mean=0 specifies the mean of the underlying normal distribution before taking the exponential function. In this case, it's set to 0, meaning that the mean of the resulting log-normal distribution will be centered at 0 after taking the exponential function.

- sigma=1 specifies the standard deviation of the underlying normal distribution before taking the exponential function. It controls the spread or width of the resulting log-normal distribution. In this case, it's set to 1, indicating a standard deviation of 1.

```python
log_normal_data = np.random.lognormal(mean=0, sigma=1, size=1000)

# Log-normal Distribution
plt.subplot(2, 2, 4)
sns.histplot(log_normal_data, kde=True)
plt.title('Log-normal Distribution')
plt.show()
```

# Gamma Distribution

The shape parameter of the gamma distribution controls its shape, skewness, and tail behavior, allowing it to represent a wide range of distributions depending on its value.

```python
gamma_data = np.random.gamma(shape=2, scale=1, size=1000)

# Gamma Distribution
plt.subplot(2, 3, 5)
sns.histplot(gamma_data, kde=True)
plt.title('Gamma Distribution')
plt.show()
```

# Second Method

```python
from scipy.stats import norm
 
mu, sigma = 0, 1  # Mean and standard deviation
x = np.linspace(-5, 5, 1000)
pdf = norm.pdf(x, mu, sigma)
 
plt.plot(x, pdf, label='Normal Distribution')
plt.title('PDF of the Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

# Uniform Distribution

```python
from scipy.stats import uniform
 
a, b = 0, 1  # Extrema of the interval
x = np.linspace(a - 0.1, b + 0.1, 1000)
pdf = uniform.pdf(x, a, b - a)
 
plt.plot(x, pdf, label='Uniform Distribution')
plt.title('Uniform Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

# Exponential Distribution

```python
from scipy.stats import expon
 
# Parameter of the exponential distribution
lambda_param = 0.5
 
x = np.linspace(0, 5, 1000)
pdf = expon.pdf(x, scale=1/lambda_param)
 
plt.plot(x, pdf, label='Exponential Distribution')
plt.title('Probability Density Function (PDF) of the Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

# Gamma Distribution

```python
from scipy.stats import gamma
 
# Parameters of the gamma distribution
alpha = 2
beta = 1
 
x = np.linspace(0, 10, 1000)
pdf = gamma.pdf(x, alpha, scale=1/beta)
 
plt.plot(x, pdf, label='Gamma Distribution')
plt.title('Probability Density Function (PDF) of the Gamma Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

## Log Normal Distribution

```python
from scipy.stats import lognorm
 
# Parameters of the log-normal distribution
mu = 0
sigma = 0.1
 
x = np.linspace(0, 3, 1000)
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))
 
plt.plot(x, pdf, label='Log-Normal Distribution')
plt.title('Probability Density Function (PDF) of the Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
