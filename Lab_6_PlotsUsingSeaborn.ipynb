```python
fig, ax = plt.subplots(1,2,figsize = (18,8))
data['Pclass'].value_counts().plot.pie(autopct= '%1.1f%%', ax= ax[0], shadow= True)
ax[0].set_title('Pclass')
sns.countplot(x='Pclass',data = data, ax= ax[1])
ax[1].set_title('Pclass')
plt.show()
```

- First create a figure (fig) and a set of subplots (ax) with 1 row and 2 columns, setting the overall figure size to (18, 8) inches.
- Then plot a pie chart on the first subplot (ax[0]) representing the distribution of passenger classes ('Pclass') in the dataset. The autopct='%1.1f%%' argument displays the percentage values on the pie chart, and shadow=True adds a shadow effect to the chart.
- Set the title of the first subplot to 'Pclass'.
- Then create a countplot (bar plot) on the second subplot (ax[1]) using Seaborn (sns). It counts the occurrences of each passenger class ('Pclass') and visualizes them as bars.
- Set the title of the second subplot to 'Pclass'.
- Finally, this command displays the complete figure with the two subplots.

Overall, this code segment provides a visual representation of the distribution of passenger classes ('Pclass') using both a pie chart and a bar plot side by side, allowing for easy comparison and analysis.

#### The Pie chart shows that:
- 1st Class passengers make up approximately 16.3% of the total.
- 2nd Class passengers make up approximately 51.5% of the total.
- 3rd Class passengers make up approximately 32.2% of the total.
#### The Bar chart confirms the information from the pie chart and provides more clarity on the number of passengers in each class:
- 1st Class: 216 passengers
- 2nd Class: 706 passengers
- 3rd Class: 531 passengers
In summary, the 'Pclass' variable in the dataset shows that most passengers traveled in 2nd Class, followed by 3rd Class, and the smallest number of passengers traveled in 1st Class.

```python
fig, ax = plt.subplots(1,2,figsize = (18,8))
data[['Gender','Pclass']].groupby(['Gender']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Gender' )
sns.countplot(x='Gender',hue='Pclass',data=data,ax=ax[1])
ax[1].set_title('Gender: Survived vs Gender')
```

- The first subplot (ax[0]) displays the average passenger class ('Pclass') for each gender category.
- The second subplot (ax[1]) shows the distribution of passenger classes ('Pclass') among different genders, with each class color-coded.
These visualizations help analyze the relationship between gender and passenger class in the dataset.

#### Bar Chart: The bar chart displays the average survival rate for each gender.
- Female passengers had a significantly higher survival rate (approximately 74.2%) compared to male passengers (approximately 18.9%).
#### Grouped Bar Chart: The grouped bar chart displays the count of survivors for each gender and passenger class 
- using seaborn's countplot function with the 'hue' parameter set to 'Pclass'. 

### A higher percentage of female passengers survived in all passenger classes compared to male passengers.
- In 1st Class, approximately 97% of female passengers survived, while only about 32% of male passengers survived.
- In 2nd Class, approximately 89% of female passengers survived, while about 16% of male passengers survived.
- In 3rd Class, approximately 50% of female passengers survived, while only about 13% of male passengers survived.


```python
fig, ax = plt.subplots(1,2,figsize = (18,8))
data[['Survived','Pclass']].groupby(['Survived']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs dead' )
sns.countplot(x='Survived',hue='Pclass',data=data,ax=ax[1])
ax[1].set_title('Gender: Survived vs Gender')
```

- The first subplot (ax[0]) displays the average passenger class (Pclass) for each survival status (Survived or Not Survived).
- The second subplot (ax[1]) shows the distribution of passenger classes (Pclass) among the survivors (Survived) and those who did not survive (Not Survived), with each class represented by a different color.
These visualizations provide insights into how passenger class relates to survival outcomes on the ship.

#### Grouped Bar Chart: The grouped bar chart displays the count of survivors for each survival status and passenger class
- using seaborn's countplot function with the 'hue' parameter set to 'Pclass'. The chart confirms the information from the bar chart and provides more clarity on the number of survivors for each survival status and passenger class:

### A higher percentage of passengers survived in higher passenger classes for both survival statuses.
- In 1st Class, approximately 63% of passengers survived, while about 37% of passengers died.
- In 2nd Class, approximately 47% of passengers survived, while about 53% of passengers died.
