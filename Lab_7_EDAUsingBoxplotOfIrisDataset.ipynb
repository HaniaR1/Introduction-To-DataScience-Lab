### Task 5: Create box plot for whole data set while keeping “species” column on x-axis. Each box plot should have a different color, its maximum and minimum limit, median point, mean point, whiskers, and outliers, if any.
#### Box Plot: 
##### A box plot, or box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the median (middle value), quartiles (dividing the data into four parts), and outliers. The box represents the interquartile range (IQR), while the whiskers extend to show the range of the data.
##### Purpose: Visualize how SepalLengthCm varies across different Species using a boxplot.
##### Customization:
- palette='Set3': Choose a color palette for aesthetics.
- width=0.5: Adjust the width of the boxes for clarity.
- linewidth=1.5: Set the width of lines outlining the boxes.
- notch=True: Display notches to indicate confidence intervals around medians.
- showmeans=True: Show means with a distinctive marker on the plot.
##### Styling:
- boxprops=dict(alpha=0.7): Control transparency of box fill.
- whiskerprops=dict(linestyle='--'): Use dashed lines for whiskers.
- capprops=dict(color='orange'): Set the color of the caps on the whiskers.
- medianprops=dict(color='blue', linestyle='-.'): Customize the style of median line.
- meanprops=dict(marker='s', markerfacecolor='red', markeredgewidth=1): Customize marker style and color for mean.
- flierprops=dict(marker='o', markerfacecolor='black', markersize=10): Customize appearance of outlier markers.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Species', y='SepalLengthCm', 
            palette='Set3', width=0.5, linewidth=1.5, 
            notch=True, showmeans=True, showcaps=True, 
            orient='v', 
            boxprops=dict(alpha=0.7), 
            whiskerprops=dict(linestyle='--'), 
            capprops=dict(color='orange'), 
            medianprops=dict(color='blue', linestyle='-.'),
            meanprops=dict(marker='s', markerfacecolor='red', markeredgewidth=1),
            flierprops=dict(marker='o', markerfacecolor='black', markersize=10))

plt.show()
```

- From this graph we can see that the specie Setosa has comparatively Smaller Sepal Length then Versicolor which is smaller than Virginica

### 2X2 Grid Boxplot
We use another method, This code creates a 2x2 grid of boxplots, each representing a different feature (SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm) across different Species. Each subplot visualizes the distribution of a specific feature among different species, enabling a comparative analysis of multiple variables simultaneously within the same figure. The use of subplots here allows for a more compact and structured presentation of the boxplot data compared to a single plot with all variables.
- The differences between the species can be observed in their measurements. For example, Iris-setosa typically has smaller sepal and petal measurements compared to Iris-versicolor and Iris-virginica. Iris-versicolor and Iris-virginica have larger sepal and petal measurements, but Iris-virginica usually has larger petal measurements compared to Iris-versicolor.
- Iris-virginica, which has the largest petal widths, also has the largest sepal widths. Similarly, Iris-setosa, which has the smallest petal widths, also has the smallest sepal widths.


```python
plt.figure(figsize=(10, 10))  # Square figure size (10x10 inches)

# Subplot 1: Boxplot for SepalLengthCm
plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Sepal Length')

# Subplot 2: Boxplot for SepalWidthCm
plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Sepal Width')

# Subplot 3: Boxplot for PetalLengthCm
plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Petal Length')

# Subplot 4: Boxplot for PetalWidthCm
plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Petal Width')

plt.show()
```

### Another way can be The code below that transforms the DataFrame (df) into a "melted" format (df_melted) to facilitate the plotting of multiple boxplots for different measurements (SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm) across Species. Each variable (Measurement) is represented as a separate boxplot, differentiated by color (hue) within each species category. Customized plot properties such as palette, line styles, transparency, and marker styles enhance visualization clarity and aesthetics. The resulting plot provides a comprehensive view of how these measurements vary across species, aiding in comparative analysis and insights.

```python
y_var = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

plt.figure(figsize=(10, 6))

# Melting the DataFrame to plot multiple boxplots
df_melted = df.melt(id_vars=['Species'], value_vars=y_var, 
                    var_name='Measurement', value_name='Value')

# Plotting the boxplot with customized properties
sns.boxplot(data=df_melted, x='Species', y='Value', hue='Measurement',
            palette='Set3', width=0.5, linewidth=1.5, 
            notch=True, showmeans=True, showcaps=True, 
            orient='v', 
            boxprops=dict(alpha=0.7), 
            whiskerprops=dict(linestyle='--'), 
            capprops=dict(color='orange'), 
            medianprops=dict(color='blue', linestyle='-.'),
            meanprops=dict(marker='s', markerfacecolor='red', markeredgewidth=1),
            flierprops=dict(marker='o', markerfacecolor='black', markersize=10))

plt.legend(title='Measurement')  # Add legend with measurement names
plt.show()
```
