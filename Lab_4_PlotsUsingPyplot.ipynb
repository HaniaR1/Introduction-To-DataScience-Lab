```python
import matplotlib.pyplot as plt
```

The matplotlib.pyplot is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. It also opens figures on your screen, and acts as the figure GUI manager.
pyplot is mainly intended for interactive plots and simple cases of programmatic plot generation

## Opening CSV File using pd

We open the CSV file by using method of pandas library called read_csv and we store it in a variable named btc. Then we print this variable 

```python
btc = pd.read_csv("BTC-USD.csv")
btc
```

The Bitcoin Data from the csv file is printed

```python
date = btc['Date']
high = btc['High']
Open = btc['Open']
close = btc['Close']
low = btc['Low']
```

We decalre variables date,high,low,Open and close to extract columns from the data stucture list. 

## Line graph between  High, Low, Open and Close price

```python
plt.figure(figsize=(10,6))
plt.plot(date,high,label='High', color='red')
plt.plot(date,low,label='Low', color='blue')
plt.plot(date,Open,label='Open', color='green')
plt.plot(date,close,label='Close', color='black')

plt.xlabel("Date")
plt.ylabel("High or Low")
plt.title("Line graph between High, Low, Open and Close price")
plt.legend()
plt.show()
```

### Description
**The following methods of the matplotlib.pyplot are called to sketch the graph**
- figure: To define size of graph
- plot: To draw line. Label and Color are used to name each line and make it of a certain color 
- high or open: Variables we are making the graph of
- xlabel: To label x axis 
- ylabel: To label y axis
- title: To Name the whole graph 
- legend: To show what colors are corresponding to what data  
- show: To display the graph

## Scatter Plot between High, Low, Open and Close price

```python
plt.scatter(high,low,color='green',alpha = 0.5)
plt.scatter(Open,close,color='red',alpha=0.5)
plt.xlabel("High")
plt.ylabel("Low")
plt.title("Scatter plot between High, Low, Open and Close price")
#plt.savefig('scatterplot.png')
plt.show()
```

### Description
**The following methods of the matplotlib.pyplot are called to sketch the graph**
- scatter: To make a scatter plot. Color is defined and transparency is set using alpha(range 0-1)
- high or open: Variables we are making the graph of
- xlabel: To label x axis 
- ylabel: To label y axis
- title: To Name the whole graph  
- show: To display the graph
- savefig: Can be used to save the graph

## Histogram for Adjoint Closing Price

```python
adj_close = btc['Adj Close']
plt.hist(adj_close,bins=20,color='green',edgecolor='black')

plt.xlabel("Adjoint Close")
plt.ylabel("Frequency")
plt.title("Histogram for Adjoint Closing price")
plt.grid(True)

plt.show()
```

### Description
**The following methods of the matplotlib.pyplot are called to sketch the graph**
- hist: To make a Histogram. This type of graph shows frequency of variable. bins us used to define number of bars, color to color these bars and edgecolor to make an outline of these bars
- adj_close: The variable we are making the graph of
- xlabel: To label x axis 
- ylabel: To label y axis
- title: To Name the whole graph  
- grid: We can turn it on or off. Grid lines can help in analysis
- show: To display the graph

## Histogram for Closing price

```python

plt.hist(close,bins=20,color='green',edgecolor='black')

plt.xlabel("Close")
plt.ylabel("Frequency")
plt.title("Histogram for Closing price")
plt.grid(True)

plt.show()
```

Similarly other histograms can also be made
